# Seaborn Distribution plots
distplot
jointplot
pairplot
rugplot
kdeplot

```python
import seaborn as sns
%matplotlib inline
```

# Import sampledataset from Seaborn

```python
df = sns.load_dataset('tips')
```

```python
df
```

# Distplot or Histogram

```python
sns.distplot(df['total_bill'],kde=False,bins=30)
```

# JointPlot
"scatter"
"reg"
"resid"
"kde"
"hex"

```python
sns.jointplot(x='total_bill',y='tip',data = df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind = 'kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind = 'reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind = 'hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind = 'resid')
```

# Pairplots

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='smoker')
```

# rugplot

```python
sns.rugplot(df['total_bill'])
```

```python
sns.regplot(data=df,x='total_bill',y='tip')
```
