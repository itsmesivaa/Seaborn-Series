# Seaborn Grids

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
import matplotlib.pyplot as plt
x = sns.PairGrid(iris)
x.map(plt.scatter)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x.map_diag(plt.hist)
x.map_upper(plt.scatter)
x.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

# FaceGrid

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
b= sns.FacetGrid(data = tips,col = 'time',row='smoker',hue='sex')
b.map(plt.scatter,'total_bill','tip').add_legend()
```

# JoinGrid

```python
y=sns.JointGrid(x='total_bill',y='tip',data=tips)
y.plot(sns.regplot,sns.distplot)
```
