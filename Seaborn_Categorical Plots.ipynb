# Seaborn Categorical Plots
* factorplot
* boxplot
* violinplot
* stripplot
* swarmplot
* barplot
* countplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

# Barplot

```python
sns.barplot(data = df,x = 'sex',y = 'total_bill')
```

# Countplot

```python
sns.countplot(data = df,x = 'sex')
```

# Boxplot

```python
sns.boxplot(data=df,x='sex',y='total_bill',hue = 'smoker')
```

# Violinplot

```python
sns.violinplot(data=df,x='day',y='total_bill')
```

# Stripplot

```python
sns.stripplot(data=df,x='day',y='total_bill',color = 'magenta',hue='sex',jitter=True)
```

# Swarmplot

```python
sns.swarmplot(data=df,x='day',y='total_bill',color = 'magenta',hue='sex')
```

```python
sns.violinplot(data=df,x='day',y='total_bill')
sns.swarmplot(data=df,x='day',y='total_bill',color = 'magenta',hue='sex')
```

# Catplot

```python
sns.catplot(data=df,x='day',y='total_bill',kind='bar')
```
