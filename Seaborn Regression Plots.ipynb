# Seaborn Regression Plots

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

# lmplot()

```python
sns.lmplot(data = tips,x='total_bill',y='tip',hue = 'sex',palette = 'rainbow')
```

# Working with Markers

```python
sns.lmplot(data = tips,x='total_bill',y='tip',hue = 'sex',palette = 'rainbow',markers=['o','v'],scatter_kws = {'s':50})
```

# Using a grid

```python
sns.lmplot(data = tips,x='total_bill',y='tip',hue = 'sex',palette = 'rainbow',col='time',row = 'sex')
```

```python
sns.lmplot(data = tips,x='total_bill',y='tip',hue = 'sex',palette = 'rainbow',col='day',row = 'sex')
```

# Aspect and Size

```python
sns.lmplot(data = tips,x='total_bill',y='tip',hue = 'sex',palette = 'rainbow',col='day',row = 'sex',aspect=0.4)
```
